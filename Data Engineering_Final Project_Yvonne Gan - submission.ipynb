### 4) Use two different parser and provide your observations (4 marks)

```python
content = BeautifulSoup(connections.content , "html.parser")
content
```

```python
content2 = BeautifulSoup(connections.content , "html5lib")
content2
```

### **Observation:**
It is a challenge to spot all the differences between "html.parser" & "html5lib" these 2 different parser at the glance. <br>
<br>
However, when having a closer look, we can see one of the differences between both parsers such as **"html.parser"** many lines of attribute were not properly closed with respective closing tag. On the other hand, **"html5lib"** is more complete as it parses pages the same way a web browser does. It is more completes with the correct close tag.

#### **Some example from the above parsed contents:** 

<u>html.parser</u>
- < link href="/favicon.ie9.ico" rel="Shortcut Icon" type="image/x-icon">
- < link href="//cdn.cnn.com/cnn/.e/img/3.0/global/misc/apple-touch-icon.png" rel="apple-touch-icon" type="image/png">


<u>html5lib</u>
- < link href="/favicon.ie9.ico" rel="Shortcut Icon" type="image/x-icon"/>
- < link href="//cdn.cnn.com/cnn/.e/img/3.0/global/misc/apple-touch-icon.png" rel="apple-touch-icon" type="image/png"/>

As you can see from the above, on the "html.parser" both attributes do not have the closing tag **'/>'**. Whereas, "html5lib" 2 listed attributes have the closing tag at the end of each line. <br>
<br>
Without the proper closing tag, it could caused an issue such as our query may grab all the attributes(lines) from the opening tag until the end of the available closing tag. These could mean many to thousand attributes compiled as One based on the open and closing tags

### 6) Identify the tag which has the headlines and sub headlines and extract them using beautiful soup. <br>(10 Marks)

```python
content2.head
```

```python
#Headline
content2.find_all("h1", class_ = "site-name")
```

```python
#Headline - title under each individual panel or commonly known for tag for the site title when it appears on single-content pages.
content2.find_all("h2" , class_ = "pane-title")
```

```python
#subheadlines
content2.find_all("h3" , class_ = "story-title")
```

```python
#Just trying to retrieve and take a closer look of the contents under h3 subheadlines tags.
content2.find_all("span", class_ = "story-headline")
```

```python
#This is to just restrive the texts format with a for loop.
for i in content2.find_all("h3" , class_ = "story-title"):
    
    print(i.get_text().rstrip().strip())
```

```python
#found a subheadline under h4
content2.find_all("h4")
```
